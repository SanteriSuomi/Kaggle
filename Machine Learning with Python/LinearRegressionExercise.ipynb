```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

data_raw = pd.read_csv("diabetes.csv")
data = data_raw.filter(["bmi"])
target = data_raw.filter(["response_interest"])
```

```
lr_model = LinearRegression()
lr_model.fit(data, target)
print(f"Equation of the regression function: ŷ = f(x) = {lr_model.coef_[0][0]}x + {lr_model.intercept_[0]}")
```

```
figure, axes = plt.subplots()
axes.scatter(data, target, color="blue", marker=".", linewidth=2, label="Data Points")
axes.set_xlabel(f"X: BMI", fontsize=11)
axes.set_ylabel(f"Y: Disease Progression", fontsize=11)
prediction = lr_model.predict(data)
axes.plot(data, prediction, color="red", linewidth=2, label="Regressor")
axes.legend(fontsize=11)
plt.show()
```

```
print(f"""Target 
{target.head()}""")
print(f"""Prediction 
{[f"{i}: {prediction[i][0]}" for i in range(5)]}""")
```

```
mse = mean_squared_error(target, prediction)
mae = mean_absolute_error(target, prediction)
r2 = r2_score(target, prediction)
print(f"""Performance Evaluation
MSE: {mse} 
MAE: {mae}
R2: {r2}""")
```
