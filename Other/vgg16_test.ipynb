```python
from google.colab import drive

from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16

import matplotlib.pyplot as plt
```

```python
 drive.mount("/content/drive")
```

```python
def preprocess_img(image):
  img_arr = img_to_array(image)
  img_arr = img_arr.reshape((1, img_arr.shape[0], img_arr.shape[1], img_arr.shape[2]))
  return preprocess_input(img_arr)
```

```python
model = VGG16()
```

```python
def predict_img(image_arr):
  y_pred = model.predict(image_arr)
  pred_res = decode_predictions(y_pred)[0][0]
  print(f"Prediction\nPredicted class: {pred_res[1]}\nPredicted probability: {int(pred_res[2] * 100)}")
```

```python
predict_img(preprocess_img(load_img("drive/My Drive/cat.jpg", target_size=(224, 224))))
```

```python
predict_img(preprocess_img(load_img("drive/My Drive/pepsimax.jpg", target_size=(224, 224))))
```
